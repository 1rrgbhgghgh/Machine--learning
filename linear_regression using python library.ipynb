```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#now import the csv file by following command
from google.colab import  files
uploaded=files.upload()
```

```python
#now import the dataset
df=pd.read_csv('Area prices.csv')
df
```

```python
#now we assume area as x ,x is independent var thats why we use [[]] bracket but y is dependent so we use []
x=df[['Area']]
y=df['Price']
x
```

```python
#Visualization
plt.scatter(x,y)
plt.xlabel('area in sqft')
plt.ylabel('prices ')
plt.title('Home prices in Dhaka')
```

```python
#no we seperate some data for train and test,for test we take 30% of data

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=1)

xtest
```

```python
#now we apply linear regression model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()

#now we train the model with our data
reg.fit(xtrain,ytrain)
ytest
```

```python
#now we predict value from testing data
reg.predict(xtest)
```

```python
#now we show a best fit line
plt.plot(df.Area,reg.predict(df[['Area']]))
plt.scatter(x,y)
plt.xlabel('area in sqft')
plt.ylabel('prices ')
plt.title('Home prices in Dhaka')
```

```python
#now we predict random value
reg.predict([[1200]])
```

```python
#show coefficient
reg.coef_
```
